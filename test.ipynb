```python
import matplotlib.pyplot as plt
import numpy as np

# Datos de crecimiento del mercado de nutricin personalizada
years = np.array([2022, 2027, 2032])
market_values = np.array([11.3, 23.3, 48.78])

# Crear grfico
plt.figure(figsize=(10, 6))
plt.plot(years, market_values, marker='o', linestyle='-', color='b')

# Configuracin del grfico
plt.title('Crecimiento del Mercado de Nutricin Personalizada (2022-2032)')
plt.xlabel('Ao')
plt.ylabel('Valor de Mercado (en miles de millones de USD)')
plt.grid(True)
plt.xticks(years)
plt.ylim(0, 50)

# Aadir anotaciones
for i, value in enumerate(market_values):
    plt.text(years[i], market_values[i] + 1, f'{market_values[i]:.2f}', ha='center')

# Guardar y mostrar grfico
plt.tight_layout()
plt.savefig("crecimiento_mercado_nutricion_personalizada.png")
plt.show()
```
